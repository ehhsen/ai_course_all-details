```python
import pandas as pd
md = {
    'cars' : ["BMW","Volvo","Ford"],
    'passings' : [1967,1997,1932]
}
myvar = pd.DataFrame(md)
print(myvar)
```

**SERIES**

```python
V = ['ARSALAN', 1,2,3]
S = pd.Series(V)
print(S)
a = [1, 7, 2]
myvar = pd.Series(a)
print(myvar[0])
```

**LABEL**

```python
# creating own labels
# label can be changed by using their index number
import pandas as pd
a = [1, 7, 2]
myvar = pd.Series(a, index = ["*", "*", "*"])
print(myvar)
```

**KEY/VALUE OBJECTS AS SERIES**

```python
CAL = {"d1" : 322, "d2":302, "d3":322}
myvar = pd.Series(CAL)
print(myvar)
```

**TASK**

```python
# accesing only 1st and 2nd elements of a series
cal = {"d1" : 322, "d2":302, "d3":322}
myvar = pd.Series(cal, index = ["d1", "d2"])
print(myvar)
```

**DATA FRAME**

```python
# COMBINATION OF ROW AND COLUMN
# 2D DATA STRUCTURE
# IN DATA SCIENCE, WE HAVE MULTI-DIMENSIONAL DATA
DATA = {
    "calories(kcal)" : [420,380,390],
    "duration(min)" : [50,40,45]
}
df = pd.DataFrame(DATA, index = ["1","2","3"])
print(df)
```

**LOCATE ROW**

```python
# LOCATING UNEXPECTED DATA
# command used loc()
import pandas as pd
DATA = {
    "calories(kcal)" : [420,380,390],
    "duration(min)" : [50,40,45]
}
df = pd.DataFrame(DATA)
print(df.loc[0])
```
