```python
# self changes private data members to public
# __init__ jaha use ho smj jao waha constructor use hua ha
class car:
  # constructor method to initialize cars attributrs
  def __init__ (self,make,model,year):
    self.make = make
    self.model = model
    self.year = year
    self.odometer_reading = 0
    # method to describe car
  def describe_car(self):
    car_info = f"{self.year} {self.make} {self.model}"
    return car_info
  # method to read odometer reading
  def read_odometer(self):
    print(f"This car has {self.odometer_reading} miles on it")
car = car("Toyota","Corolla",2022)
print(car.describe_car())
car.read_odometer()
```

```python
class calcul():
  def __init__ (self,a,b):
    # The __init__ method should not return any value
    self.a = a
    self.b = b
  def sum(self):
    return self.a + self.b
  def sub(self):
    return self.a - self.b
  def mul(self):
    return self.a * self.b
  def div(self):
    return self.a / self.b
a,b = int(input("enter a: ")),int(input("enter b: "))
c = calcul(a,b)
op = input("enter operator +,-,*,/ : ")
if op == "+":
  print(c.sum())
elif op == "-":
  print(c.sub())
elif op == "*":
  print(c.mul())
elif op == "/":
  print(c.div())
```
