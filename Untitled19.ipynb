**TASK: REPLACING NULL VALUES IN CALORIES**

```python
import pandas as pd
df = pd.read_csv('data.csv')
df["Calories"].fillna(130, inplace = True)
print(df.to_string())
print(df.info())
```

```python
# replace using mean median mode
import pandas as pd
df = pd.read_csv('data.csv')
x = df["Calories"].mean()          #average
df["Calories"].fillna(x, inplace = True)
print(x)
print(df.to_string())
```

```python
# repacing null values with median
import pandas as pd
df = pd.read_csv('data.csv')
y = df["Duration"].median()
df["Duration"].fillna(x, inplace = True)
print(y)
print(df.to_string())
```
