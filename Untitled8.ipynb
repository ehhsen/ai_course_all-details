```python
# dictionary
d = {
    "brand: ": "Ford",
    "model: ": "Mustang",
    "year: ": "1964"
   }

print(d)
print(d["brand: "])
print(d.get("model: "))
d[1]= "honda"
print(d)
print(d.pop("brand"))
```
