```python
# arbitrary arguments
def av(*num):
  s =0
  for i in num:
    s += i
  print(s/len(num))
av(10,15)
```

```python
# result calculator
def calc(a):
  if a >= 90 and a<=100:
    print("A")

  elif a >= 80 and a<90:
    print("B")

  elif a >= 70 and a<80:
    print("C")
  elif a >= 60 and a<70:
    print("D")
  elif a >= 50 and a<60:
    print("E")
  elif a>=40 and a<50:
    print("F")
  else:
    print("invalid")


  x = float(input("enter percentage : "))
  calc(x)
```

```python
# result calculator
def calc(inp):
  if inp >= 90 and inp <=100:
    print("A")

  elif inp >= 80 and inp <90:
    print("B")

  elif inp >= 70 and inp <80:
    print("C")
  elif inp >= 60 and inp <70:
    print("D")
  elif inp >= 50 and inp <60:
    print("E")
  elif inp>=40 and inp<50:
    print("F")
  else:
    print("invalid")

y = float(input("enter percentage : "))
calc(y)
```

```python
def my_dec(func):
  def wrapp():
    print("Something is happening before the function")
    func()
    print("Something is happening after the function")
  return wrapp
@my_dec
def say_hello():
  print("hello")
say_hello()
```
