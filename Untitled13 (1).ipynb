```python
class Employee():
    def __init__(self,name,age):
        self.name = name
        self.age = age
    def showdet(self):
        print("name of employee is", self.name)
class programmar(Employee):
    def showlang(self):
        print("The default language is python")
e1 = programmar("ahsan",19)
e1.showdet()
e1.showlang()

```

```python
class super_class():
    def __init__ (self,name,age,height,qual):
        self.name = name
        self.age = age
        self.height = height
        self.qual = qual
class sub_class(super_class):
    def details(self,siblings):
        self.siblings = siblings
    def show(self):
        print(f"my name is {self.name}, age is {self.age},\n height is {self.height}, qualification is{self.qual}\n and siblings are {self.siblings}")

e3 = sub_class()
e3.name = "Ahsan"
e3.age = 19
e3.siblings = 2
e3.height = 5.8
e3.qual = "intermediate"
e3.show()

# e3.details()
e3.show()
```

```python
class f():
    f_name = ""
    def fa(self):
        print(self.f_name)
class m():
    m_name= ""
    def mo(self):
        print(self.m_name)
class son(f,m):
    def parents(self):
        print("Father: ",self.f_name)
        print("Mother: ",self.m_name)
s1 = son()
s1.f_name = "Abc"
s1.m_name = "xyz"
s1.parents()
```

```python
class car():
  def __init__ (self,model,brand):
    self.model = model
    self.brand = brand
  def move(self):
    print("Drive")
class boat():
  def __init__ (self,model,brand):
    self.model = model
    self.brand = brand
  def move(self):
    print("Sail")
car1 = car("ford", "Mushtang")
boat1 = boat("Ibiza", "touring 20")
for x in (car1, boat1):
  x.move()
  print(x)
```
