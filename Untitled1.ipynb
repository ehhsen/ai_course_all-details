```python
x = float(input("Enter your percentage "))
match x:
  case _ if x>80 and x<=100:
    print("A")
  case _ if x>70 and x<=79:
    print("B")
  case _ if x>60 and x<=69:
    print("C")
  case _ if x>50 and x<=59:
    print("D")
  case _ if x>40 and x<=49:
    print("E")
  case _ if x>0 and x<39:
    print("F")
  case _ :
    print("Invalid")
```
