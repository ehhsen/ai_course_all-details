```python
# spot test
# QUESTION 1 : WRITE A PROGRAM TO FIND PRIME NUMBER
x= int(input("Enter a number: "))              #input
if x > 1:                                    # number = 1 or less than 1 is not prime
    for i in range(2, x):                    #checking for multiples of x from 2 to x
        if (x % i) == 0:                      #checking if x is divisible by i
            print(x, "is not a prime number")
            break
    else:
        print(x, "is a prime number")
else:
    print(x, "is not a prime number")
```
