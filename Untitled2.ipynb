```python
#POSITIVE SLICING - LEFT TO RIGHT
name = "mango"
print(name[0:4])# full
print(name[ :4])# default start
print(name[0: ])# default nd
print(name[ : ])# default start and end
print(name[2:4])
print(name[:])
print(name[ :5])
```

positive slicing
variable_name [start:end]
end== it gives end-1 character
end_value +1 gives whole string


```python
#NEGATIVE SLICING - RIGHT TO LEFT
# execution start from -1
# right value should be ggreater than left
name = "mango"
print(name[ :-1])
print(name[ :-4])
print(name[-4:-1])
print(name[-1:-4]) # right value should be ggreater than left
```

```python
#string methods
n= "lad harry"
print(n.upper())                 #all uppper case
print(n.lower())                 #all lower case
print(n.capitalize())            #first letter capital
print(n.title())                 #first letter capital of each word
print(n.swapcase())              #swap case
print(n.replace(" ","    "))     #change any thing or character or whole word/string
print(n.rstrip())                #remove white spaces from right- can only rremove last character rstrip("r") xx   rstrip("y") true
print(n.lstrip())                #remove white spaces from left
print(n.strip())                 #remove white spaces from both left and right
print(n.find("h"))               #find character or word
print(n.count("a"))              #count character or word
print(n.split(" "))              #split string into list
first, second = n.split(" ")
print(first, type(first))
print(second, type(second))

```
