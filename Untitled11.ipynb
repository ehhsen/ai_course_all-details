**ERORS AND EXCEPTIONS**

*   **TRY**
*   **EXCEPT**






*   **FINALLY/ELSE**




```python
# # syntax
# try:
#   # code that may cause error
# except:
#   # code which wil run in case of exception
# else:
#   # code which will be executed if try is true
# finally:
#   # code which will be executed regardless of exception
try:
  10/0

except ZeroDivisionError :
   print(" cant be divided by zero ")
else:
  print("no error")
finally:
  print("this statement is executed independently")
```

**ZeroDivisionError exception**

```python
try:
  nume = int(input("enter numerator: "))
  denum = int(input("enter denominato: r"))
  print(nume/denum)
except:
  print("division by zero or invalid input, enter integers only")

```

**INDEX ERROR EXCEPTIONS**

```python
list_1 =[1,2,3,4,5]
try:
  a = int(input("enter index: "))
  print(list_1[a])
except:
  print("index out of range")
```

**PRACTISING MULTIPLE EXCEPTOINS**

```python
# multiple exceptions
try:
  nume = int(input("enter numerator: "))
  denum = int(input("enter denominator:"))
  print(nume/denum)
except ZeroDivisionError :
  print(" cant be divided by zero ")
except ValueError:
  print("invalid input, enter integers only")
```

**FIND EVEN WITH EXCEPTIONS**

```python
try:
  num = int(input("enter number: "))
  assert num%2 == 0
except AssertionError:
  print("number is not even")
else :
  print("number is even")
  print(" reciprocal of input is ", 1/num)
```

**FIND ODD WITH EXCEPTIONS**

```python
try:
  num = int(input("enter number: "))
  assert num%2 != 0
except AssertionError:
  print("number is even")
else :
  print("number is odd")
  print("reciprocal of input is ", 1/num)
```
